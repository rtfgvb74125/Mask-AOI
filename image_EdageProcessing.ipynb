```python
import pandas as pd
import numpy as np
import cv2
import os
```

```python
label_list = ['error_ear' , 'error_iron' , 'error_solder' , 'lint' , 'NG' , 'noearing' , 'noiron' , 'normal']

in_path = 'E:\\mask_data\\insta\\validation\\'

out_path = 'E:\\mask_data\\insta_Edge_Nomove_BG\\validation\\'

# BG_img = cv2.imread('E:\\mask_data\\background.jpg')
# BG_img = cv2.cvtColor(BG_img, cv2.COLOR_BGR2GRAY)
# BG_img = cv2.resize(BG_img , (960,540))
# sobelX = cv2.Sobel(BG_img, cv2.CV_64F, 1, 0)
# sobelY = cv2.Sobel(BG_img, cv2.CV_64F, 0, 1)
# sobelX = np.uint8(np.absolute(sobelX))
# sobelY = np.uint8(np.absolute(sobelY))
# sobelCombined_BG = cv2.bitwise_or(sobelX, sobelY)

# in_file = os.listdir(in_path)
for label in label_list:
#     print(label)
    path = in_path+label
#     print(path)
    in_file = os.listdir(in_path+label)
    for file in in_file:
#         print(file)
        img = cv2.imread(os.path.join(path,file))
        img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        img = cv2.resize(img , (960 , 540))
        sobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0)
        sobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1)
        sobelX = np.uint8(np.absolute(sobelX))
        sobelY = np.uint8(np.absolute(sobelY))
        sobelCombined = cv2.bitwise_or(sobelX, sobelY)
#         no_BG = cv2.subtract(sobelCombined,sobelCombined_BG)
        cv2.imwrite(out_path+label+'\\'+file,sobelCombined)
        print(out_path+label+'\\'+file)
```
